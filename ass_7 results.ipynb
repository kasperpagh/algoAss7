```python
%%sql select forename,surname,driverid,sum(rank) as wins from drivers 
join results using (driverid) where results.rank = 1 
group by driverid 
order by wins desc
```

```python
%%sql select drivers.driverref, constructors.name, count(constructorid) as "driven x times" from drivers 
join results using (driverid) 
join constructors using (constructorid)
group by driverid,constructors.name
order by "driven x times" desc
limit 50
```

```python
%%sql select drivers.driverref, constructors.name, count(constructorid) as "driven x times" from drivers 
join results using (driverid) 
join constructors using (constructorid)
join status using (statusid)
where status like 'Finished'
group by driverid,constructors.name
order by "driven x times" desc
limit 50
```

```python
%%sql select driverid, raceid, sum(milliseconds) as "time spent in pitstops" from drivers
join pitstops using (driverid)
group by driverid,raceid
order by raceid asc
limit 50
```

```python
%%sql create VIEW race_pitstops as
select driverid, raceid, sum(milliseconds) as "time spent in pitstops", count(driverid) as "number of pitstops" from drivers
join pitstops using (driverid)
group by driverid,raceid
order by raceid asc
```

```python
%sql select * from race_pitstops limit 10
```

```python
%%sql select drivers.driverref, constructors.name, count(constructorid) as "driven x times",
p2.tp as "time in pitstop" from drivers 
join results using (driverid) 
join constructors using (constructorid)
join status using (statusid)
join (select driverid,sum("number of pitstops") as np,sum("time spent in pitstops") as tp from race_pitstops group by driverid order by driverid) as p2 using (driverid)
where status like 'Finished'
group by driverid,constructors.name,p2.tp
order by "time in pitstop" desc
limit 50
```

```python
%%sql select drivers.driverref, constructors.name, count(constructorid) as "driven x times",
p2.tp as "time in pitstop",p2.tp/p2.np as "time pr stop" from drivers 
join results using (driverid) 
join constructors using (constructorid)
join status using (statusid)
join (select driverid,sum("number of pitstops") as np,sum("time spent in pitstops") as tp from race_pitstops group by driverid order by driverid) as p2 using (driverid)
where status like 'Finished'
group by driverid,constructors.name,"time in pitstop","time pr stop"
order by "time pr stop" asc
limit 50
```
